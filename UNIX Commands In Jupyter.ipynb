### Counting most frequent words

```python
!sed -e 's/ /\'$'\n/g' < $filename | sort | uniq -c |sort -nr | head -15
```

Head Stops After 15 Lines, Hence Sort Command Is Sending The Last Two Errors

#### Writing Output To A File

```python
!sed -e 's/ /\'$'\n/g' < $filename | sort | uniq -c |sort -nr | head -15 > count_words.txt
```

```python
!cat count_words.txt
```

#### Plotting By Importing WordCounts Into Python

```python
import matplotlib.pyplot as plt
import csv
```

```python
xTicks  = []
y =[]

with open('count_words.txt', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter =' ')
    for row in plots:
        y.append(int(row[-2]))
        xTicks.append(str(row[-1]))
        
# removing the count of spaces in 1st line
y = y[1:]
xTicks = xTicks[1:]

#plotting
x = range(len(y))
plt.figure(figsize = (10,10))
plt.xticks(x, xTicks, rotation =90) #xlabel with 90 degree angle
plt.plot(x,y,'*')
plt.show()


```
